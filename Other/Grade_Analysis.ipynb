```python
# Make a copy
df1 = df.copy(deep = True)
df1
```

```python
# Converting column headings to lowercase
df1.columns = df1.columns.str.lower()
df1.columns
```

```python
# Replacing whitespaces with underscores on the columns
df1.rename(columns = lambda col_name:col_name.replace(" ", "_"), inplace = True)
df1
```

```python
# Remove index 0
df1 = df1.drop(0, axis = 0)
```

```python
# Resetting the index again to start at 0
df1.reset_index(drop = True)
```

```python
# Slicing through columns to get the quiz columns
quizzes = []
for quiz in df1.columns:
    if "quiz" in quiz:
        quizzes.append(quiz)
main_quizzes= quizzes[:5]
print(type(main_quizzes))
main_quizzes

#Alt method: List Comprehension
# quizzes = [quiz for quiz in df1.columns if "quiz" in quiz]
# main_quizzes = quizzes[:5]
```

```python
df_quizzes = df1[main_quizzes]
df_quizzes['student'] = df1[['student']]
reordered_cols = ['student'] + [col for col in df_quizzes.columns if col != 'student']
df_quizzes = df_quizzes[reordered_cols]
df_quizzes
```

```python
# Getting labs from df1.columns
labs = []
for lab in df1.columns:
    if "codegrade" in lab:
        labs.append(lab)
len(labs)
```

```python
# Ordering the columns in our new dataframe df_labs
column_order = ['student'] + labs
# df_labs = df1[labs].copy()
df_labs = df1[column_order]
df_labs
```

```python
df_labs.reset_index(drop = True)
df_labs
```

```python
# Assigning a variable to all the labs where no one attempted
unattempted_labs = df_labs.columns[df_labs.isna().sum() == 30]
```

```python
#Dropping the unattempted_labs
df_labs = df_labs.drop(columns = unattempted_labs)
df_labs
```
