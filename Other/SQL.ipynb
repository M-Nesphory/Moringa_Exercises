```python
import sqlite3
import pandas as pd
db = sqlite3.connect("chinook.db")
pd.read_sql("""
SELECT * 
FROM sqlite_master
WHERE type = "table";
""", db)
```

```python
pd.read_sql("""
SELECT *
FROM customers;
""", db)
```

```python
pd.read_sql("""
SELECT FirstName, LastName, country
FROM customers
WHERE country = "USA"
ORDER BY FirstName DESC
LIMIT 10;
""", db)
```

```python
pd.read_sql("""
SELECT Country, COUNT(Country) AS num_customers
FROM Customers
GROUP BY Country
ORDER BY num_customers DESC;
""", db)
```

```python
pd.read_sql("""
SELECT * 
FROM invoices;
""", db)
```

```python
pd.read_sql("""
SELECT *, SUM(Total) AS total_invoiced
FROM invoices
WHERE Total > 2
GROUP BY BillingCountry
HAVING total_invoiced > 10
ORDER BY total_invoiced DESC;
""", db)
```

```python
pd.read_sql("""
SELECT *
FROM Customers
WHERE State IS NULL;
""", db)
```
