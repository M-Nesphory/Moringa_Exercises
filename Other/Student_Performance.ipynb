```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("SAP-4000.csv")
df.head()
```

```python
# Gender and Region vs Exam Score
sns.barplot(data = df, 
            x = 'Gender', 
            y = 'Exam_Score', 
            hue = 'Region',
           errorbar = None);
```

```python
# Hours Studied vs Exam Score
sns.scatterplot(y ='HoursStudied/Week', x='Exam_Score',
                hue = 'Gender',
                palette =['#347433','#FFC107'],
             data = df);
# y: y axis
# x: x axis
# hue: add a categorical dimension
# palette: Change colors of the hue
# data: dataframe
```

```python
df['Gender'].value_counts(normalize = True)
```

```python
# Exam Score vs Attendance
plt.figure(figsize=(15, 15))
sns.scatterplot(x = 'Exam_Score', y = 'Attendance(%)',
               data = df,
               size = 'HoursStudied/Week',
               sizes = (20,200));
```
