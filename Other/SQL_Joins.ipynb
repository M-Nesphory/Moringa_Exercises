```python
import sqlite3
import pandas as pd
import pandasql
db = sqlite3.connect("flights.db")

pd.read_sql("""
SELECT *
FROM sqlite_master
WHERE type = 'table';
""", db)
```

```python
pd.read_sql("""
SELECT *
FROM airports;
""", db)
```

```python
pd.read_sql("""
SELECT *
FROM airlines;
""", db)
```

```python
pd.read_sql("""
SELECT *
FROM routes;
""", db)
```

```python
# Which countries have the highest number of active airlines (Top 10)
pd.read_sql("""
SELECT country, COUNT() AS num_airlines
FROM airlines
WHERE active = "Y"
GROUP BY country
ORDER BY num_airlines DESC
LIMIT 10;
""", db)
```

```python
# Which countries have the highest number of inactive airlines?
# Return only those with more than 10
pd.read_sql("""
SELECT country, COUNT() AS num_inactive_airlines
FROM airlines
WHERE active = "N"
GROUP BY country
    HAVING num_inactive_airlines >=10
ORDER BY num_inactive_airlines DESC;
""", db)
```

```python
# Run a query that returns the number of airports per time zone
# Each row should have a number of airports and a time zone
pd.read_sql("""
SELECT timezone, COUNT() AS num_airports
FROM airports
GROUP BY timezone
ORDER BY num_airports DESC;
""", db)
```

```python
# Which airline has the most routes listed?
pd.read_sql("""
SELECT a.name, COUNT() AS total_routes
FROM airlines AS a
JOIN routes AS r
    ON a.id = r.airline_id
GROUP BY a.name
ORDER BY total_routes DESC
LIMIT 10;
""", db)
```
