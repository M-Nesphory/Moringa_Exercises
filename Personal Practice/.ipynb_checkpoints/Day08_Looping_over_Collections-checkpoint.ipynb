**For Loops**

For loops are used to iterate through list elements one by one.

They can iterate through any data type.

```python
zero_to_three = [0,1,2,3]
for num in zero_to_three:
    print(num)
# The for loop allows us to iterate through the list
```

**Syntax Breakown**

The for loop takes in two arguments.

**num** is the variable name we are assigning to an element

**zero_to_three** is the collection we are iterating over

Every for loop needs to end the first line in a colon. It indicates the start of a code block; which is indicated by indentation.

```python
counter = 0
for num in zero_to_three:
    counter+=1
    print("This is iteration", counter)
    print(num)
print("End of loop")
```

**Looping over a dictionary**

variable *key* prints out the keys in the dictionary, while variable *value* prints out the values.

While looping through a dict, ensure the dictionary name has an *.items()* method at the end of it.

```python
countries_cities = {
    "Croatia": "Zagreb",
    'USA': 'District of Columbia',
    'Argentina':'Buenos Aires',
    'France': 'Paris',
    'Brazil': 'Brasilia',
    'Japan': 'Tokyo',
    'Vietnam': 'Hanoi',
    'Kenya': 'Nairobi'
}
for key,value in countries_cities.items():
    print(f"Country: {key}\nCity: {value}")
```

When iterating using a for loop, sometimes we do not know the extent of the number.

We can use the range object to show the extent of the looping

**Syntax:** range(a,b) 

where a= starting point and b = ending point excluding the end

```python
for num in range(0, 10):
    print(num)
    # 10 is excluded
```

```python
countries_cities = {
    "Croatia": "Zagreb",
    'USA': 'District of Columbia',
    'Argentina':'Buenos Aires',
    'France': 'Paris',
    'Brazil': 'Brasilia',
    'Japan': 'Tokyo',
    'Vietnam': 'Hanoi',
    'Kenya': 'Nairobi'
}
for country, city in countries_cities.items():
    print(country)
```

```python
import pandas as pd
file_name = './cities.xlsx' # Assignes the excel sheet to a variable
travel_df = pd.read_excel(file_name) # Reads the excel file and converts to travel_df
cities = travel_df.to_dict('records') # Converts the DataFrame into a dictionary using to_dict() method
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

x_values = [cities[0]['City'], cities[1]['City'], cities[2]['City']]
y_values = [cities[0]['Population'], cities[1]['Population'], cities[2]['Population']]

plt.bar(x_values, y_values)
plt.ylabel('Population')
plt.title('City Populations')

plt.show()
```

```python
city_names = []
# If starting at 0, the range object would not need the starting int arguement, just the end
for index in range(len(cities)):
    city_names.append(cities[index]['City'])
city_names
```

```python
names_and_ranks = []

for index in range(len(cities)):
    num = index + 1
    names_and_ranks.append(f"{num}. {city_names[index]}")
names_and_ranks
```

```python
city_populations = []

for index in range(len(cities)):
    city_populations.append(cities[index]['Population'])
city_populations
```

```python
plt.bar(names_and_ranks, city_populations)
plt.xticks(rotation='vertical')
plt.ylabel('Population')
plt.title('City Populations')
plt.show()
```

```python
city_areas = []

for index in range(len(cities)):
    city_areas.append(cities[index]['Area'])
city_areas
```

```python
plt.bar(names_and_ranks, city_populations)

plt.ylabel('Population')
plt.xlabel('Cities')
plt.title('City Populations')
plt.xticks(rotation='vertical')

plt.show()
```

```python
plt.bar(names_and_ranks, city_areas)
plt.ylabel('Area')
plt.xlabel('Cities')
plt.title('City Areas')
plt.xticks(rotation='vertical')

plt.show()
```
