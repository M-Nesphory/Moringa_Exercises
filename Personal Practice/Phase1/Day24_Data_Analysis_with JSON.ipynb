```python
# Let's see how botanical families are represented in our data
# Build a dictionary that counts the plants by family
# Use the family_common_names for this
```

```python
plant_dict = {}
for plant in plants:
    plant_fam_name = plant['family_common_name']
    if plant_fam_name not in plant_dict:
        plant_dict[plant_fam_name] = 1
    else:
        plant_dict[plant_fam_name] += 1
plant_dict
```

```python
# Plants with no family names
plant_dict[None]
```

```python
# Remove all values with no families attached to them
# Call them orphans hehehehehehehehe
fams_clean = {fam: num for fam,num in plant_dict.items() if fam != None}
```

```python
# Let us also remove those families with counts < 10
fams_clean = {fam: num for fam,num in fams_clean.items() if num >=10}
fams_clean
```

```python
# Make a bar chart of the numbers
fig, ax = plt.subplots(figsize=(20, 12))
ax.bar(list(fams_clean.keys()), list(fams_clean.values()))
ax.set_title("Families by Number", fontsize = 30, color = 'r')
plt.xticks(rotation = 80, fontsize = 20)
plt.yticks(fontsize = 20);
```

```python
# How many synonymous names are listed for each plant
```

```python
num_sym = [len(plant['synonyms']) for plant in plants]
num_sym
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(range(1000), num_sym)
ax.set_title('Number of Synonyms');
```

```python
#Find the outlier
outlier = [plant['common_name'] for plant in plants if len(plant['synonyms']) > 250]
outlier
```
