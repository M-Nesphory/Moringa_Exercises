**Scatter Plot**

Sometimes called Correlation plots

Used to show the realtionship between two variables

```python
import numpy as np
x = np.linspace(0,10,30)
# np.linspace generates data starting from 0 to 10 and creates 30 new elements.
# Calculate sin(x) and save in a new array
y = np.sin(x)
print(y)
```

```python
plt.scatter(x,y, label = "Function: sin x") # Label - provides more info
plt.title("Scatter Plot") # Defines a title
plt.legend() # Adds legend info
plt.show() # display the plot
```

```python
plt.figure(figsize=(10,6)) # Specify width and height of the plot in inches

plt.scatter(x, y, label = "y = sin(x)" )

# Set x and y axes labels
plt.xlabel('X Values') # X axis label
plt.ylabel('Y Values') # Y axis label

plt.title('Scatter Plot in Matplotlib')
plt.legend()
plt.show()
```

**Bar Graphs**

Shows comparisons across data. Can be plotted vertically or horizontally.

Bar graphs are useful as they show how data is distributed across categories

```python
# Copied as is from Colab
# Set seed for reproducibility
np.random.seed(100)

# Generate the x-axis variable as 10 categories using numpy's arange function
x = np.arange(10)

# For y-axis, generate 10 random quantities
y = np.random.randn(10)
```

```python
plt.figure(figsize=(10,6))

# Use the bar() function to create a plot using the above values of x and y. Add a label.
plt.bar(x, y, label='Sample Data')

plt.xlabel('X Values - Categories') # X-axis label
plt.ylabel('Y Values - Quantities') # Y-axis label

plt.title('Bar Plot in Matplotlib') # Title
plt.legend() # I think this is unneccesary if it is only one data

# Output the final plot
plt.show()
```

**Histograms**

Histograms help you detect the frequency distribution of a data set. Helps to show if skewness and outliers.
The 
X-axis describes the bins/groups they are put in, while the y-axis describes the frequency of the occurrence


```python
# Copied as is
# Set seed for reproducability
np.random.seed(100)

# Generate a data set of 200 retirement age values
x = 5*np.random.randn(200) + 65

#Plot the histogram with hist() function
plt.hist(x, bins = 10, edgecolor='black')
"""
x axis
number of bins/groups on x axis
color of the column edges
"""

plt.xlabel('Retirement Age') # x-axis label
plt.ylabel('Frequency of Values') # y-axis label
plt.title('Histograms in Matplotlib') # title of histogram
plt.show() # Output the plot
```

```python
years = range(1975, 2000)
# Create a numpy array of 25 values from 0 - 1000
data = np.linspace(0,1000,25)
print(data)
```

**Lineplot**

Used to represent a trend. X-axis indicates a time measure

Look like a scatter plot but each point is connected using a trend line

```python
# Create the plot
plt.subplots()

# Use plot() function to create a plot using above values
plt.plot(years, data)

# Add a legend to the plot
plt.legend(["Sample Data"]);
'''
Notice the semicolon at the end of the last line. 
If this is not included, the location in memory of the last object 
to be created in the graph can be displayed before the graph is displayed.
'''
```

**Lineplot labelling**

- plt.set_xlabel('') - define x-axis label
- plt.set_ylabel('') - define y-axis label
- plt.set_title('') - define plot title
- plt.legend('') - adds a legend. You can also define the location of the legend:
  
      - plt.legend('', loc =1) - Upper right corner
      - plt.legend('', loc =2) - Upper left corner
      - plt.legend('', loc =3) - Lower left corner
      - plt.legend('', loc =4) - Lower right corner

To plot multiple sets of data on the same axes, call multiple plotting methods on the same ax object

```python
# Adding fake data
other_data = data - (data % 150) + 50

# Creating the plot
fig, ax = plt.subplots()

# Plotting both sets of data
ax.plot(years, data)
ax.plot(years, other_data)

# Setting axes labels
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")

# Title
ax.set_title("This is a title")

# Legend in the lower right corner
ax.legend(['Sample Data','Other Data'], loc = 4);
```

You can also plot more than one type of axes on the same axes

```python
x1 = [1, 4, 6, 8]
y1 = [10, 15, 27, 32]
x2 = [0.5, 2.2, 4.2, 6.5]
y2 = [21, 19, 9, 26]

fig,ax = plt.subplots()

# Line plot
ax.plot(x1,y1)

# Scatter plot
ax.scatter(x2,y2)

# Legend
ax.legend(['Dataset 1', 'Dataset 2']);
```

```python
# Generate data
x = np.arange(101)
y = x**2

# Create the plot
fig, ax = plt.subplots()

# Draw a line graph
ax.plot(x,y)

# Set title
ax.set_title("Line Plot with Default Axes");
```

For aesthetics, one can 'zoom in' or 'zoom out' of a plot by setting the limits of the axis

- set_xlim(min,max)
- set_ylim(min,max)

```python
# Create plot, draw graph, set title
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("Zoomed Out")

# Set the limits of x and y to "zoom out"
ax.set_xlim(min(x)-15, max(x)+15)
ax.set_ylim(min(y)-1500, max(y)+1500);

# Outputs - Leaves a sort of padding near the edges line plot
```

```python
# Create plot, draw graph, set title
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("Zoomed In")

# Set the limits of x and y to "zoom in"
ax.set_xlim(80, 100)
ax.set_ylim(6000, 10000);

# No boundary along the edges of the line plot
```

You can also change the scales and numbering of the ticks on the axes themselves. This depends on the granularity the user needs.

- .set_xticks()
- .set_yticks()

Wrong axis ticks can make for poor graphs

```python
# Create plot, draw graph, set title
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("More Ticks on x-axis, Fewer Ticks on y-axis")

# Customize the x and y axis ticks so x-axis has 11 and y-axis has 5
xticks = np.linspace(start=min(x), stop=max(x), num=11)
yticks = np.linspace(start=min(y), stop=max(y), num=5)
ax.set_xticks(xticks)
ax.set_yticks(yticks);
```

**Customizing Line Styles**

- color - color of the plot
- linewidth - thickness of the line
- linstyle - line pattern
- marker - point pattern (mostly for scatter plots)
- etc


```python
# Create the plot
fig, ax = plt.subplots()

# Set the limits of x and y axes
ax.set_xlim(0, 9), ax.set_ylim(5,35)

# Generate a line plot with custom styling
ax.plot(x1, y1, color='cyan', linewidth=4, linestyle = '-.')

# Draw a scatter plot on same axes with custom styling
ax.scatter(x2, y2, color='red', marker='x')

# Add a legend
ax.legend(["Dataset 1", "Dataset 2"]);
```

**Summary of line styles in Matplotlib**

```python
# Set up data and plot
x = np.arange(0, 10)
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+4, color="red", linewidth=1.00)
ax.plot(x, x+6, color="red", linewidth=2.00)

# possible linestyle options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+10, color="green", lw=3, linestyle='-')
ax.plot(x, x+12, color="green", lw=3, ls='-.')
ax.plot(x, x+14, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+18, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+22, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+24, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+26, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+28, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+32, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+34, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+36, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+38, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

We can also create multiple axes (i.e. multiple subplots) within a single figure by specifying the nrows and/or ncols arguments.

```python
# Note axes
fig, axes = plt.subplots(figsize = (11, 3), ncols = 3)
# Creates 3 side-by-side axes within the figure
```

```python
fig, axes = plt.subplots(figsize= (3,11), nrows = 3)
```

axes is a collection of axes objects stored in a list-like data structure.

We can access individual axes object as follows:

```python
fig, axes = plt.subplots(figsize =(11, 3), ncols = 3)
axes[0].set_facecolor('blue') # First subplot
axes[1].set_facecolor('orange') # Second subplot
axes[2].set_facecolor('red') # Third subplot
```

**Plotting with Subplots**

Let's re-draw the above combined line plot and scatter plot in two different subplots.

```python
# Plot creation
fig, axes = plt.subplots(figsize = (11,4), ncols = 2)

# Setting limits
axes[0].set_xlim(0,9), axes[0].set_ylim(7,35)
axes[1].set_xlim(-1,8), axes[1].set_ylim(5,30)

# Plotting line graph on one subplot
axes[0].plot(x1,y1, color = "lightblue", linewidth = 3, linestyle = '-.')

# Plotting scatter plot on other graph
axes[1].scatter(x2, y2, color = 'red', marker = 'x')

# Title to the figure
fig.suptitle("Plots Across Two Subplots", fontsize =24, x = 0.44)

# Legend
fig.legend(labels = ['Dataset 1','Dataset 2'], loc = (.68,.88));
```
