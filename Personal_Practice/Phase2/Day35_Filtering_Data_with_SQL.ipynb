The `WHERE` clause is used to filter data in a database. It is used to check if rows match to a specific condition.

It is done using `=`.

````python
# Select all employees whose lastname = Patterson
pd.read_sql("""
SELECT *
FROM employees
WHERE lastName = "Patterson";
""", conn)
````

````python
# Select order details where rounded price = 30
pd.read_sql("""
SELECT *, CAST(ROUND(priceEach) AS INT) AS rounded_price
FROM orderDetails
WHERE rounded_price = 30;
""", conn)
````

````python
# Select all orders that were placed in January
pd.read_sql("""
SELECT *, strftime("%m", orderDate) AS month
FROM orders
WHERE month = "01"
""", conn)
````

````python
# Select all orders that were shipped late
# shippedDate after requiredDate
pd.read_sql("""
SELECT *, julianday(shippedDate) - julianday(requiredDate) AS days_late
FROM orders
WHERE days_late > 0;
""", conn)
````

## Conditional Operators in SQL

In all of the above queries, we used the `=` operator to check if we had an exact match for a given value. However, what if you wanted to select the order details where the price was at least 30 dollars? Or all of the orders that don't currently have a shipped date?

We'll need some more advanced conditional operators for that.

Some important ones to know are:

* `!=` and `<>` ("not equal to")
  * Similar to `not` combined with `==` in Python
* `>` ("greater than")
  * Similar to `>` in Python
* `>=` ("greater than or equal to")
  * Similar to `>=` in Python
* `<` ("less than")
  * Similar to `<` in Python
* `<=` ("less than or equal to")
  * Similar to `<=` in Python
* `AND`
  * Similar to `and` in Python
* `OR`
  * Similar to `or` in Python
* `BETWEEN`
  * Similar to placing a value between two values with `<=` and `and` in Python, e.g. `(2 <= x) and (x <= 5)`
* `IN`
  * Similar to `in` in Python
* `LIKE`
  * Uses wildcards to find similar strings. No direct equivalent in Python, but similar to some Bash terminal commands.

`BETWEEN` keyword is used to select all values within a range.

Used together with the `AND` keyword

````python
# Select all cats between 1 and 3 years old
pd.read_sql("""
SELECT *
FROM cats
WHERE age BETWEEN 1 AND 3;
""", db)
````

`NULL` in SQL represents missing values

We use the `IS` keyword to check for null values and `IS NOT` to check the inverse

````python
# Select pets whose owners are unknown
pd.read_sql("""
SELECT *
FROM cats
WHERE owner_id IS NULL;
""", db)
````

````python
# Select pets whose owners are known
pd.read_sql("""
SELECT *
FROM cats
WHERE owner_id IS NOT NULL;
""", db)
````

The `LIKE` operator is used with wildcards to identify values that have a specific pattern. The `%` symbol is the most used wildcard

If we want to get all values that begin with `M` we shall write `M%`. The query will return all values that begin with M.

- Ending with M - `%M`
- Has M - `%M%`

The other wildcard is `_` which means exactly one character with any value

For example, if we wanted to select all cats with four-letter names where the second letter was "a", we could use `_a__`.

````python
# Select all cats that start with M
pd.read_sql("""
SELECT *
FROM cats
WHERE name LIKE 'M%';
""", db)
````

````python
pd.read_sql("""
SELECT *
FROM cats
WHERE name LIKE "_a__";
""", db)
````

## Note on `SELECT`

We are now familiar with this syntax:

```sql
SELECT name
  FROM cats;
```

However, you may not know that this can be written like this as well:

```sql
SELECT cats.name
  FROM cats;
```

Both return the same data.

SQLite allows us to explicitly state the `tableName.columnName` you want to select. This is particularly useful when you want data from two different tables.

Imagine you have another table called `dogs` with a column containing all of the dog names:

```sql
CREATE TABLE dogs (
	id   INTEGER PRIMARY KEY,
	name TEXT
);
```

```sql
INSERT INTO dogs (name)
VALUES ("Clifford");
```


If you want to get the names of all the dogs and cats, you can no longer run a query with just the column name.
`SELECT name FROM cats,dogs;` will return `Error: ambiguous column name: name`.

Instead, you must explicitly follow the `tableName.columnName` syntax.
```sql
SELECT cats.name, dogs.name
  FROM cats, dogs;
```

You may see this in the future. Don't let it trip you up!

````python
# Select just the name and color of each planet
pd.read_sql("""
SELECT pl.name, pl.color
FROM planets AS pl
""", db)
````

````python
# Select all columns for each planet whose num_of_moons is 0
pd.read_sql("""
SELECT *
FROM planets
WHERE num_of_moons = 0;
""", db)
````

````python
# Select name and mass of each planet whose name has exactly 7 letters
pd.read_sql("""
SELECT name, mass
FROM planets
WHERE length(name) = 7;
""", db)
````

````python
# Select all columns for each planet whose mass is greater than 1.00
pd.read_sql("""
SELECT *
FROM planets
WHERE mass > 1;
""", db)
````

````python
# Select the name and mass of each planet whose mass is less than or equal to 1.00
pd.read_sql("""
SELECT name, mass
FROM planets
WHERE mass <= 1;
""", db)
````

````python
# Select the name and mass of each planet whose mass is between 0 and 50
pd.read_sql("""
SELECT name, mass
FROM planets
WHERE mass BETWEEN 0 AND 50;
""", db)
````

````python
#Select all columns for planets that have 
# at least one moon and a mass less than 1.00
pd.read_sql("""
SELECT *
FROM planets
WHERE num_of_moons >= 1 AND mass < 1;
""", db)
````

````python
# Select the name and color of planets 
# that have a color containing the string "blue"
pd.read_sql("""
SELECT *
FROM planets
WHERE color LIKE "%blue%";
""", db)
````

````python
# Select the count of planets that don't have rings as planets_without_rings
pd.read_sql("""
SELECT COUNT(rings) AS planet_without_rings
FROM planets
WHERE rings = 0;
""", db)
````

````python
# Select the name of all planets, 
# along with a value has_rings that returns 
# "Yes" if the planet does have rings, and "No" if it does not
pd.read_sql("""
SELECT name, 
    CASE rings
    WHEN 1 THEN "Yes"
    WHEN 0 THEN "No"
    END AS has_rings
FROM planets;
""", db)
````
