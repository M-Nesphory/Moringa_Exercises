## ORDER BY

Normally, a database is ordered in the way they way input in the database
`Order By` is used to sort the order of the returned rows

It can be alphabetically or numerically

**Syntax:** ORDER BY column_name (sort_order);

Sort Order:
- ASC - Ascending Order (default)
- DESC - Descending order

```python
# Order products by productName 
# (ascending order as default)
pd.read_sql("""
SELECT *
FROM products
ORDER BY productName ASC;
""", db)
```

```python
# Order products by productName in descending order
pd.read_sql("""
SELECT *
FROM products
ORDER BY productName DESC;
""", db)
```

```python
# Sorting by the length of the productDesciption column
pd.read_sql("""
SELECT productName, length(productDescription) AS desc_length
FROM products
ORDER BY desc_length DESC;
""", db)
```

You can also sort by multiple columns by putting the two columns in the ORDER BY clause.

The first column is ordered first then the second. Order matters

```python
pd.read_sql("""
SELECT productVendor, productName, MSRP
FROM products
ORDER BY productVendor, productName;
""", db)
```

As a rule of thumb to figure out which column should be ordered first, the column that has fewer unique values should be ordered first.

Use the `DISTINCT` keyword with `COUNT` to figure out the number of unique values in each column

```python
pd.read_sql("""
SELECT COUNT (DISTINCT productVendor) AS count_PV,
    COUNT (DISTINCT productName) AS count_PN
FROM products;
""", db)
# Order by productVendor first
```

Sometimes when ordering, the order might be wonky due to difference in datatype. E.g, when ints are stored as strings. Python/SQL wont realize they are numbers and thus will not order them correctly. It is thus advisable to use `CAST` in the `ORDER BY` clause to get numbers

```python
# Without casting
pd.read_sql("""
SELECT productName, quantityInStock
FROM products
ORDER BY quantityInStock;
""", db)
```

```python
# With casting
pd.read_sql("""
SELECT productName, quantityInStock
FROM products
ORDER BY CAST(quantityInStock AS INT);
""", db)
```

## LIMIT

`Limit` is used to determine the  *maximum* number of records that you want to return.

**Syntax:** LIMIT number

Works like `.head()` in Python

`LIMIT` is very powerful when used with `ORDER BY`

```python
pd.read_sql("""
SELECT *
FROM products
LIMIT 5;
""", db)
```

```python
# What are the top 10 orders with the longest comments
pd.read_sql("""
SELECT *, length(comments)
FROM orders
ORDER BY length(comments) DESC
LIMIT 10;
""", db)
```

```python
pd.read_sql("""
SELECT *, length(comments)
FROM orders
WHERE status = "Cancelled"
ORDER BY length(comments) DESC
LIMIT 10;
""", db)
```

```python
# LIMIT can also be used with datetime data
# Who were the first five unique customers to make an order
pd.read_sql("""
SELECT DISTINCT customerNUmber, orderDate
FROM orders
ORDER BY orderDate
LIMIT 5;
""", db)
```

```python
# Of the orders that have not been shipped and have not been canceled, 
# what are the 10 newest based on order date?
pd.read_sql("""
SELECT *
FROM orders
WHERE shippedDate IS NOT NULL AND status != "Cancelled"
ORDER BY orderDate DESC
LIMIT 10;
""", db)
```

```python
# What is the order that took the longest to fulfill, 
# and how long did it take?
pd.read_sql("""
SELECT *, 
    julianday(shippedDate) - julianDay(orderDate) AS days_to_fulfill
FROM orders
WHERE shippedDate IS NOT NULL
ORDER by days_to_fulfill DESC
LIMIT 1;
""", db)
```

# Ordering and Limiting Data Lab

```python
dogs = sqlite3.connect(r"C:\Users\nrmmw\Documents\Flatiron\Repos\Phase_2\dsc-filtering-and-ordering-lab\dogs.db")
pd.read_sql("""
SELECT *
FROM dogs;
""", dogs)
```

```python
# Select the name and breed for all female dogs
pd.read_sql("""
SELECT name, breed
FROM dogs
WHERE gender = "F";
""", dogs)
```

```python
# Select the number of dogs that do not have a name
pd.read_sql("""
SELECT COUNT() AS num_dogs
FROM dogs
WHERE name IS NULL;
""", dogs)
```

```python
# Select the names of all dogs that contain the double letters ff or oo
pd.read_sql("""
SELECT name
FROM dogs
WHERE name LIKE "%oo%" OR name LIKE "%ff%";
""", dogs)
```

```python
# Select the names of all dogs listed in alphabetical order. 
# Notice that SQL lists the nameless dog first.
pd.read_sql("""
SELECT name
FROM dogs
ORDER BY name;
""", dogs)
```

```python
# Select the name and breed of only the hungry dogs 
# and list them from youngest to oldest
pd.read_sql("""
SELECT name, breed
FROM dogs
WHERE hungry = 1
ORDER BY age ASC;
""", dogs)
```

```python
# Select the oldest dog's name, age, and temperament
pd.read_sql("""
SELECT name, age, temperament
FROM dogs
ORDER BY age DESC
LIMIT 1;
""", dogs)
```

```python
# Select the name and age of the three youngest dogs
pd.read_sql("""
SELECT name, age
FROM dogs
ORDER BY age ASC
LIMIT 3;
""", dogs)
```

```python
# Select the name and breed of the dogs 
# who are between five and ten years old, 
# ordered from oldest to youngest
pd.read_sql("""
SELECT name, breed, age
FROM dogs
WHERE age BETWEEN 5 AND 10
ORDER BY age DESC;
""", dogs)
```

```python
# Select the name, age, and hungry columns for hungry dogs 
# between the ages of two and seven. 
# This query should also list these dogs in alphabetical order.
pd.read_sql("""
SELECT name, age, hungry
FROM dogs
WHERE hungry = 1 AND 
    age BETWEEN 2 AND 7
ORDER BY name ASC;
""", dogs)
```
