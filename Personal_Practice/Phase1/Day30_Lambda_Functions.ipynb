Lmabda functions are used to write throw-away of anonymous functions on the go. They are a quick and concise way to write functions

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\nrmmw\Documents\Flatiron\dsc-lambda-functions\Yelp_Reviews.csv",
                index_col = 0)
df.head()
```

```python
# Lambda function to count the words in each review
df['text'].map(lambda txt: len(txt.split())).head()
```

```python
#Lambda can also accept conditionals if chained
df['text'].map(lambda review: "Good" if any([word in review.lower() for word in ['awesome','love','good','great']]) else 'Bad').head()
```

```python
# Say we want to select the year from a date
df.date.map(lambda x: x[:4]).head()
```

```python
# Lambdas are useful within the sort() function too
names = ['Miriam Marks','Sidney Baird','Elaine Barrera','Eddie Reeves','Marley Beard',
         'Jaiden Liu','Bethany Martin','Stephen Rios','Audrey Mayer','Kameron Davidson',
         'Carter Wong','Teagan Bennett']
# Sorting by last name
sorted(names, key = lambda name: name.split()[1])
```

# Lambda Functions Lab

Use a lambda function to create a new column called `'stars_squared'` by squaring the stars column.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\nrmmw\Documents\Flatiron\dsc-lambda-functions\Yelp_Reviews.csv",
                index_col = 0)
df.head()
```

```python
df['star_squared'] = df['stars'].apply(lambda x: x**2)
df.head()
```

Select the month from the date string using a lambda function.

```python
df['date'].map(lambda x: x[5:7]).head()
```

What is the average number of words for a yelp review?

```python
df['text'].map(lambda review: len(review.split())).mean()
```

Create a new column for the number of words in the review

```python
df['text_length'] = df['text'].map(lambda x:len(x.split()))
df.head()
```

```python
# Rewrite the following function as a lambda function
def rewrite_as_lambda(value):
    if len(value.split()) < 50:
        return 'Short'
    elif len(value.split()) < 80:
        return 'Medium'
    else:
        return 'Long'
# Hint: nest your if, else conditionals

df['Review_length'] = None
```

```python
df['Review_length'] = df['text'].map(lambda words: "Short" if len(words.split()) < 50 else 'Medium' if len(words.split()) < 80 else "Long")
df.head(10)
```

```python
# This is honestly simpler
df['Review_length'] = df['text'].map(rewrite_as_lambda)
df.head(5)
```
