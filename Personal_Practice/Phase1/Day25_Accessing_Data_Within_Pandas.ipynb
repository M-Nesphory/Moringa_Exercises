### Selcting DataFrame Information
- `.iloc` - used for integer-location based indexing
- `.loc` - selecting by label/index

## .iloc

```python
#iloc
# Selecting the 4th row
df.iloc[3]
```

```python
# Slicing rows
# iloc[a:b] - a included, b excluded
df.iloc[5:8]
```

```python
df.iloc[:, 3:7]
# : selects all rows
# the comma separates the rows from the cols
# 3:7 selects the the third - sixth columns
```

```python
# Slicing both rows and columns
df.iloc[50:61,5:10]
#Voila
```

## .loc

```python
# .loc selects columns based on their row index and the col name
# Selecting all rows under the magnesium column
df.loc[:,'magnesium']
# df['magnesium'] (Also works)
```

```python
df.loc[7:16, 'magnesium']
```

```python
#df.loc can also be used to select rows 
#that fulfill a specific condition
df.loc[df['alcohol'] < 12]
#df[df['alcohol'] < 12] also works
```

However, the .loc attribute is useful if you'd only want the color intensity for the wines with an alcohol percentage below 12. You can obtain the result as follows:

```python
df.loc[df['alcohol'] < 12, ['color_intensity']]
```

Most of these methods and selectors are also applicable to Pandas Series. See how you can convert a one-column DataFrame into a Pandas Series:

```python
col_intensity = df['color_intensity']
type(col_intensity)
```

```python
col_intensity[0:3]
```

```python
col_intensity[col_intensity > 8]
```

Imagine that for some reason, you're not interested in the color intensity values for color intensities above 10, and simply want to set all color intensities to 10 when they are bigger than 10. You can use a selector method and then assign it a new value, just like this:

```python
df.loc[df['color_intensity'] > 10, 'color_intensity'] = 10
```

```python
df[df["color_intensity"] >=10]
```

Now imagine that we want to create a new column named, "shade" which has a value, "light" when the color_intensity is below 7, and, "dark" when the intensity is > 7. This can be done as follows:

```python
# Creating New columns
df.loc[df['color_intensity'] > 7, 'shade'] = 'dark'
df.loc[df['color_intensity'] <= 7, 'shade'] = 'light'
```

Now, let's try to find out how many countries participated in the 1986 World Cup.

```python
df_1986 = df.loc[df['Year'] == 1986, (["Home Team Name", "Away Team Name"])]
df_1986
```

```python
#tolist() - converts it into a list
countries = df_1986['Home Team Name'].tolist()
countries.extend(df_1986['Away Team Name'].tolist())
participants = set(countries)
print(f'There were {len(participants)} participants in the 1986 World Cup')
participants
```

Imagine that, for some reason, we simply want Korea listed as one entry, so we want to replace every "Home Team Name" and "Away Team Name" entry that contains "Korea" to simply "Korea". In the same way, we want to change the columns "Home Team Initials" and "Away Team Initials" to NSK (North & South Korea) instead of "KOR" and "PRK

```python
# Changing Team Names
df.loc[df['Home Team Name'] == 'Korea Republic', 'Home Team Name'] = 'Korea'
df.loc[df['Home Team Name'] == 'Korea DPR', 'Home Team Name'] = 'Korea'
df.loc[df['Away Team Name'] == 'Korea Republic', 'Away Team Name'] = 'Korea'
df.loc[df['Away Team Name'] == 'Korea DPR', 'Away Team Name'] = 'Korea'
```

```python
df[df['Home Team Name'] == 'Korea Republic']
```

```python
# Channging Team Initials
df.loc[df['Home Team Initials'] == 'KOR', 'Home Team Initials'] = 'NSK'
df.loc[df['Home Team Initials'] == 'PRK', 'Home Team Initials'] = 'NSK'
df.loc[df['Away Team Initials'] == 'KOR', 'Away Team Initials'] = 'NSK'
df.loc[df['Away Team Initials'] == 'PRK', 'Away Team Initials'] = 'NSK'
```
